# simulating a social Network

```python
DG=nx.gn_graph(7,seed=25)
for line in nx.generate_edgelist(DG,data=False): print(line)
```

```python
print(DG.node[0])
```

```python
DG.node[0]['name']='Alice'
print DG.node[0]
```

```python
DG.node[1]['name']='Bob'
DG.node[2]['name']='Ankit'
DG.node[3]['name']='Annu'
DG.node[4]['name']='Jenifer'
DG.node[5]['name']='Susy'
DG.node[6]['name']='Jony'
```

```python
DG.add_nodes_from([(0,{'age':24}),(1,{'age':25}),(2,{'age':15}),(3,{'age':19}),(4,{'age':32}),(5,{'age':10}),(6,{'age':24})])
print DG.node[0]
```

```python
DG.node[0]['gender']='f'
DG.node[1]['gender']='m'
DG.node[2]['gender']='m'
DG.node[3]['gender']='f'
DG.node[4]['gender']='f'
DG.node[5]['gender']='f'
DG.node[6]['gender']='m'
```

```python
nx.draw_circular(DG,node_color='bisque',with_labels=True)
```

```python
label_dict={}
for i in range(7):
    label_dict[i]=DG.node[i]['name']
nx.draw_circular(DG,node_color='bisque',with_labels=True,labels=label_dict)    
```

```python
G=DG.to_undirected()
nx.draw_spectral(G,labels=label_dict,node_color='bisque',with_labels=True)
```
