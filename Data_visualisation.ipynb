# Time series plot

```python
df_time=pd.read_csv('SuperstoreSales.csv',index_col='Order Date',parse_dates=True)
df_time.head()
```

```python
df_time['Order Quantity'].plot()
```

```python
df_time2=df_time.sample(n=100,random_state=25,axis=0)
plt.xlabel('Order Date')
plt.ylabel('Order Quantity')
plt.title('SuperStore Sales')
df_time2['Order Quantity'].plot()
```

## Adding an extra dimension

```python
cars_df=pd.DataFrame(cars.iloc[:,[1,3,4,6]].values,columns=['mpg','disp','hp','wt'])
#iloc only allows lists to be parsed (no tuples)
```

```python
cars_target=cars.iloc[:,9].values
cars_target
```

```python
target_names=[0,1]
cars_df['group']=pd.Series(cars_target,dtype='category')
sb.pairplot(cars_df,hue='group', palette='hls')
```
