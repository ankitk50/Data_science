```python
from pylab import *
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,5,10)
y=x ** 2
```

```python
figure()
```

```python
plot(x,y,'r')
xlabel('x')
ylabel('y')
title('title')
show()
```

```python
subplot(1,2,1)
plot(x,y,'r--')
subplot(1,2,2)
plot(x,y,'b*--');
show()
```

```python
fig= plt.figure()
axes1=fig.add_axes([.1,.1,.8,.8])#left, bottom , width , height(range 0 to 1)
axes2=fig.add_axes([.2,.5,.4,.3])#inlet axis
axes1.plot(x,y,'g*--')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('Title_outer')
axes2.plot(y,x,'b--')
axes2.set_xlabel('x')
axes2.set_ylabel('y')
axes2.set_title('Title_inner')
show()
```
